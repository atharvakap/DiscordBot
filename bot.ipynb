```python
def process_image(img, user, font, event, name):
  # print(user)
  # print(name)
  width, height = img.size
  rgb_im = img.convert('RGB')
  draw = ImageDraw.Draw(rgb_im)
  text_width, text_height = draw.textsize(user, font=font)
  new_width = (width/2) - (text_width/2)
  draw.text(xy=(new_width, 650),text='{}'.format(user),fill=(0,0,0),font=font)
  rgb_im.save('/content/{}.pdf'.format(name + "_" + event))
```

```python
import py_avataaars as pa

def avatar_features(style, skin_color, top_type, hair_color, facial_hair_type, facial_hair_color, mouth_type, eye_type, eyebrow_type, accessories_type, clothe_type, clothe_color, graphic_type, hat_color):
  option_skin_color = skin_color
  option_style = style
  option_top_type = top_type
  option_hair_color = hair_color
  option_hat_color = hat_color
  # option_hair_type = input(hair_type)
  option_facial_hair_type = facial_hair_type
  option_facial_hair_color = facial_hair_color
  option_mouth_type = mouth_type
  option_eye_type = eye_type
  option_eyebrow_type = eyebrow_type
  option_accessories_type = accessories_type
  option_clothe_type = clothe_type
  option_clothe_color = clothe_color
  option_clothe_graphic_type = graphic_type

  avatar = pa.PyAvataaar( 
             style=eval('pa.AvatarStyle.%s' % option_style),
    skin_color=eval('pa.SkinColor.%s' % option_skin_color),
    top_type=eval('pa.TopType.SHORT_HAIR_SHORT_FLAT.%s' % option_top_type),
    hair_color=eval('pa.HairColor.%s' % option_hair_color),
    # hat_color=eval('pa.ClotheColor.%s' % option_hat_color),
    facial_hair_type=eval('pa.FacialHairType.%s' % option_facial_hair_type),
    facial_hair_color=eval('pa.HairColor.%s' % option_facial_hair_color),
    mouth_type=eval('pa.MouthType.%s' % option_mouth_type),
    eye_type=eval('pa.EyesType.%s' % option_eye_type),
    eyebrow_type=eval('pa.EyebrowType.%s' % option_eyebrow_type),
    nose_type=pa.NoseType.DEFAULT,
    accessories_type=eval('pa.AccessoriesType.%s' % option_accessories_type),
    clothe_type=eval('pa.ClotheType.%s' % option_clothe_type),
    clothe_color=eval('pa.Color.%s' % option_clothe_color),
    clothe_graphic_type=eval('pa.ClotheGraphicType.%s' %option_clothe_graphic_type),
    hat_color = eval('pa.Color.%s' %option_hat_color)
  )

  # rendering the avatar in png format
  avatar.render_png_file("avatar.png")
```
